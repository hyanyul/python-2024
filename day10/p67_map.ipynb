````python
# 지도 표시, 지도 확인 시 구글맵 참조
#https://map.google.com
map = folium.Map(location=(35.177769, 129.198844))
````

````python
#map: 줌값 필요함
map = folium.Map(location=(37.240860, 131.867265), zoom_start=17, zoom_control=True)   #zoom_start: 0~18, 보통 15정도를 기본으로 사용
map #(!=print(map))
````

````python
map.save('mymap.html')
map
````

- 지도 타일 셋 지정
    - OpenStreetMap(기본), OpenStreetMap.HOT, OpenStreetMap.BZH(글자 없이 지도만 나옴), OpenTopoMap(로딩 오래 걸림)
    - CartoDB Positron, CartoDB PositronNoLabels, CartoDB Voyager, CartoDB Dark Matter(CartoDB 그룹)
    - CyclOSM

````python
#타일 셋 지정
map = folium.Map(location=[37.240860, 131.867265], tiles='CartoDB Voyager', zoom_start=17)
map
````

````python
map = folium.Map(location=[35.178508, 129.199443], tiles='CartoDB Voyager', zoom_start=15)

folium.Marker(location=[35.178508, 129.199443], tooltip='송정 해수욕장', popup='서핑 명소').add_to(map)
popup = folium.Popup('<pre><a href="https://www.starbucks.co.kr/">스타벅스</a>로 오세요</pre>', min_width=200, max_width=300)
folium.Marker(location=[35.180762, 129.202324], tooltip='스타벅스 부산송정비치점', popup=popup, icon=folium.Icon(color='darkgreen', icon='heart')).add_to(map)

map
````
